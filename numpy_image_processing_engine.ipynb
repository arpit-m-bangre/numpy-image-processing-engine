## 🚀 Features Implemented
- Grayscale conversion
- Brightness adjustment

- Blur using convolution

---
## 📥 Load Image
Load an image and convert it into a NumPy array.
---


```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load image
img = Image.open("myphoto.jpg")

# Convert to NumPy array
img = np.array(img)

print("\n--- Image Loaded Successfully ---\n")
print("Shape:", img.shape)
print("Data type:", img.dtype)

# Display image
plt.figure(figsize=(6, 4))
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
print("\n--- Pixel Value Range ---\n")
print("Min pixel:", img.min())
print("Max pixel:", img.max())
```

---
## 🎨 Grayscale Conversion
Convert RGB image into a single-channel grayscale image by averaging pixel values.
---


```python
# Convert to grayscale
gray = np.mean(img, axis=2)

print("\n--- Grayscale Image ---\n")
print("Grayscale shape:", gray.shape)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Grayscale")
plt.imshow(gray, cmap='gray')
plt.axis("off")
plt.show()
```

```python
print("\n--- Grayscale Pixel Range ---\n")
print("Min:", gray.min())
print("Max:", gray.max())
```

---
## 🔆 Brightness Adjustment
Increase or decrease pixel intensity values to control image brightness.
---


```python
# Increase brightness
bright = img + 50

# Prevent overflow (IMPORTANT)
bright = np.clip(bright, 0, 255)

print("\n--- Brightness Adjusted Image (+50) ---\n")
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Increased Brightness")
plt.imshow(bright.astype(np.uint8))
plt.axis("off")
plt.show()
```

```python
# Decrease brightness
dark = img - 50
dark = np.clip(dark, 0, 255)

print("\n--- Brightness Adjusted Image (-50) ---\n")
plt.figure(figsize=(6, 4))
plt.title("Decreased Brightness")
plt.imshow(dark.astype(np.uint8))
plt.axis("off")
plt.show()
```

---
## 🎛️ Contrast Enhancement
Stretch pixel values around the midpoint to improve visibility of features.
---


```python
# Contrast factor
factor = 1.5

# Apply contrast formula
contrast = 128 + factor * (img - 128)

# Clip values
contrast = np.clip(contrast, 0, 255)

print("\n--- Contrast Enhanced Image ---\n")
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Enhanced Contrast")
plt.imshow(contrast.astype(np.uint8))
plt.axis("off")
plt.show()
```

```python
print("\n--- Different Contrast Levels ---\n")
plt.figure(figsize=(15,4))
for i, f in enumerate([0.5, 1.0, 1.5, 2.0]):
    contrast = 128 + f * (img - 128)
    contrast = np.clip(contrast, 0, 255)

    plt.subplot(1, 4, i+1)
    plt.imshow(contrast.astype(np.uint8))
    plt.title(f"Contrast: {f}")
    plt.axis("off")

plt.show()
```

## 🔄 Spatial Transformations
Perform horizontal/vertical flipping and crop specific regions of the image.
---


```python
# Horizontal flip (mirror)
flip_h = img[:, ::-1]

print("\n--- Horizontally Flipped Image ---\n")
plt.figure(figsize=(6, 4))
plt.imshow(flip_h)
plt.title("Horizontal Flip")
plt.axis("off")
plt.show()
```

```python
# Vertical flip
flip_v = img[::-1, :]

print("\n--- Vertically Flipped Image ---\n")
plt.figure(figsize=(6, 4))
plt.imshow(flip_v)
plt.title("Vertical Flip")
plt.axis("off")
plt.show()
```

```python
# Crop (adjust values based on your image)
crop = img[50:200, 50:200]

print("\n--- Cropped Image ---\n")
plt.figure(figsize=(6, 4))
plt.imshow(crop)
plt.title("Cropped Region")
plt.axis("off")
plt.show()
```

## 🌫️ Blur using Convolution
Apply a smoothing filter by averaging neighboring pixels using a kernel.
---


```python
from scipy.signal import convolve2d

# Convert to grayscale for simplicity
gray = np.mean(img, axis=2)

# Kernel
kernel = np.ones((3,3)) / 9

# Apply convolution
blur = convolve2d(gray, kernel, mode='same')

print("\n--- Blurred Image ---\n")
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("Grayscale Original")
plt.imshow(gray, cmap='gray')
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Blurred")
plt.imshow(blur, cmap='gray')
plt.axis("off")
plt.show()
```

## 🔍 Edge Detection (Sobel Filter)
Detect edges by identifying rapid intensity changes using gradient filters.
---


```python
from scipy.signal import convolve2d

# Convert to grayscale (important)
gray = np.mean(img, axis=2)

# Sobel X kernel
sobel_x = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])

# Apply convolution
edges_x = convolve2d(gray, sobel_x, mode='same')

print("\n--- Edge Detection (Sobel X) ---\n")
plt.figure(figsize=(6, 4))
plt.imshow(edges_x, cmap='gray')
plt.title("Sobel X Edges")
plt.axis("off")
plt.show()
```

```python
# Sobel Y kernel
sobel_y = np.array([
    [-1,-2,-1],
    [ 0, 0, 0],
    [ 1, 2, 1]
])

edges_y = convolve2d(gray, sobel_y, mode='same')

# Combine edges
edges = np.sqrt(edges_x**2 + edges_y**2)

print("\n--- Edge Detection (Combined Sobel) ---\n")
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("Original Grayscale")
plt.imshow(gray, cmap='gray')
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Combined Edges")
plt.imshow(edges, cmap='gray')
plt.axis("off")
plt.show()
```

```python
edges = np.clip(edges, 0, 255)

print("\n--- Clipped Edges ---\n")
plt.figure(figsize=(6, 4))
plt.imshow(edges, cmap='gray')
plt.title("Clipped Edges")
plt.axis("off")
plt.show()
```

```python
print("\n--- Thresholded Edges (>100) ---\n")
plt.figure(figsize=(6, 4))
plt.imshow(edges > 100, cmap='gray')
plt.title("Strong Edges")
plt.axis("off")
plt.show()
```
